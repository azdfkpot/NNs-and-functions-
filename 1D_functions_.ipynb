```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense

# Define the neural network architecture
model = Sequential()
model.add(Dense(50, input_dim=1, activation='tanh'))
model.add(Dense(50, activation='tanh'))
model.add(Dense(1))

# Prepare the data for training
x_train = np.linspace(2, 10, 1000).reshape(-1, 1)
y_train = np.cos(2*x_train) + x_train*np.sin(3*x_train) + np.sqrt(x_train)

# Train the neural network
model.compile(loss='mse', optimizer='adam')
model.fit(x_train, y_train, epochs=300, batch_size=32, verbose=0)

# Evaluate the trained neural network
x_test = np.linspace(2, 10, 1000).reshape(-1, 1)
y_test = np.cos(2*x_test) + x_test*np.sin(3*x_test) + np.sqrt(x_test)
y_pred = model.predict(x_test)

# Plot the exact curve and the neural network curve
plt.plot(x_test, y_test, label='Exact')
plt.plot(x_test, y_pred, label='Approximation by NN')
plt.legend()
plt.show()
```
